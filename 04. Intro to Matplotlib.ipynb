# Matplotlib

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

### plot 

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y);
```

### Plot $y=x^2+5$ line

```python
x = np.linspace(-10, 10, 100)
y = x**2 + 5

plt.plot(x, y);
```

### Plot $âˆš(a*x)$

```python
def sqrt(x, a):
    return np.sqrt(a*x)
```

```python
x = np.linspace(2, 10, 100)
y = sqrt(x, a=1)

plt.plot(x, y);
```

```python
x = np.linspace(1, 10, 100)

plt.plot(x, sqrt(x, a=1))
plt.plot(x, sqrt(x, a=2))
plt.plot(x, sqrt(x, a=3))
plt.plot(x, sqrt(x, a=4))
plt.plot(x, sqrt(x, a=5));
```

```python
plt.plot(x, sqrt(x, a=2))
plt.plot(x, -sqrt(x, a=2));
```

### Plot `cos()` function

```python
x = np.linspace(0, 10, 1000)
y = np.cos(x)

plt.plot(x, y);
```

## Plot properties

### linestyles

- https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html

```python
x = np.linspace(1, 5, 10)

plt.plot(x, x, linestyle='solid')
plt.plot(x, np.log(x), linestyle='dotted')
plt.plot(x, x*3, linestyle='dashdot')
plt.plot(x, x*4, linestyle='dashed')
```

### color

```python
x = np.linspace(1, 5, 100)

plt.plot(x, x, color='g')
plt.plot(x, np.log(x), color='blue')
plt.plot(x, np.sin(x), color='k')
plt.plot(x, np.cos(x), color='red')
```

## marker

- https://matplotlib.org/stable/gallery/lines_bars_and_markers/marker_reference.html

```python
plt.plot(x, x+0, marker='x', markersize=5)
plt.plot(x, x+1, marker='_')
plt.plot(x, x+2, marker='3')
plt.plot(x, x+3, marker='x', markersize=10)
```

### title and xlabel, ylabel

```python
x = np.linspace(1, 5, 10)
plt.title('Plot sinh and -sinh')
plt.plot(x, np.sinh(x), color='r', label='sinh', marker='4')
plt.plot(x, -np.sinh(x), color='k', label='-sinh', marker=9)
plt.legend()
plt.xlabel('x')
plt.ylabel('sinh')
plt.grid();
```

## scatter plot

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(100)
y = np.random.rand(100)

plt.scatter(x, y)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')

plt.show()
```
