# Hypothesis function

If we have one feature (also known as a univariate linear regression), the formula of the line (also known as the **hypothesis function**) in linear regression can be simplified to:

**y_hat = β₀ + β₁x** 

where:

- y_hat is the dependent variable or model output (also known as the response variable)
- x is the independent variable or input (also known as the predictor variable)
- β₀ is the y-intercept (also known as the bias term)
- β₀ determines the point at which the line intersects the y-axis. A larger value of β₀ shifts the entire line up, while a smaller value of β₀ shifts the entire line down.
- 
- β₁ is the coefficient (or weight) of the independent variable x
- β₁ determines the slope of the line. A larger value of β₁ makes the line steeper, while a smaller value of β₁ makes the line flatter.
- β₀ and β₁ are model parameters

```python
plt.scatter(x_train, y_train);

x = np.linspace(0, 300, 100)

b0 = 5
b1 = 0.080

y_hat = b0 + b1*x
plt.plot(x, y_hat, 'r')
```

## Example

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
from sklearn.datasets import make_regression
```

```python
x, y = make_regression(n_samples=500, n_features=1, noise=0.5)
```

```python
x.shape, y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Scale the data using StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Train a Linear Regression model
reg = LinearRegression()
reg.fit(X_train_scaled, y_train)
```

### Evaluating model performance (mean squared error)

```python
y_pred = reg.predict(X_test_scaled)
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse:.2f}")
```

### Evaluating model performance (R-squared)

```python
reg.score(X_test_scaled, y_test)
```

```python
reg.coef_, reg.intercept_
```

### Plot the predicted and true values with the hypothesis function

```python
fig, ax = plt.subplots(figsize=(8, 6))
ax.scatter(y_test, y_pred)
ax.plot([y.min(), y.max()], [y.min(), y.max()], 'k--', lw=2)

ax.set_xlabel('True Values')
ax.set_ylabel('Predictions')
ax.set_title('Predicted vs. True Values')
plt.show()
```
