# Imports

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis
from sklearn.metrics import accuracy_score, classification_report
from mlxtend.plotting import plot_decision_regions
```

# Load the Iris dataset

```python
iris = load_iris()
X = iris.data[:, 1:3]  # We'll only use the first two features for visualization purposes
y = iris.target
```

# Split the dataset into training and testing sets

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

# Initialize LDA and QDA classifiers

```python
lda = LinearDiscriminantAnalysis()
qda = QuadraticDiscriminantAnalysis()
```

# Fit the models

```python
lda.fit(X_train, y_train)
qda.fit(X_train, y_train)
```

# Predictions

```python
y_pred_lda = lda.predict(X_test)
y_pred_qda = qda.predict(X_test)
```

# Compute accuracy

```python
accuracy_lda = accuracy_score(y_test, y_pred_lda)
accuracy_qda = accuracy_score(y_test, y_pred_qda)
```

# Classification report

```python
report_lda = classification_report(y_test, y_pred_lda)
report_qda = classification_report(y_test, y_pred_qda)
```

# Plot decision boundaries

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plot_decision_regions(X_test, y_test, clf=lda, legend=2)
plt.title(f'LDA Decision Boundary with Accuracy of {accuracy_lda:.2}')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')

plt.subplot(1, 2, 2)
plot_decision_regions(X_test, y_test, clf=qda, legend=2)
plt.title(f'QDA Decision Boundary with Accuracy of {accuracy_qda:.2}')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')

plt.tight_layout()
plt.show()
```

# Print results

```python
print("LDA Accuracy:", accuracy_lda)
print("LDA Classification Report:\n", report_lda)
print("\nQDA Accuracy:", accuracy_qda)
print("QDA Classification Report:\n", report_qda)
```
