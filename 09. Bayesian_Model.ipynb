```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import BernoulliNB, GaussianNB, MultinomialNB, ComplementNB
from mlxtend.plotting import plot_decision_regions
```

```python
class NaiveBayesModels:
    def __init__(self):
        self.models = {
            'Gaussian Naive Bayes': None,
            'Multinomial Naive Bayes': None,
            'Bernoulli Naive Bayes': None,
            'Complement Naive Bayes': None
        }

    def load_data(self):
        iris = load_iris()
        self.X = iris.data[:, :2] 
        self.y = iris.target
        self.X_train, self.X_test, self.y_train, self.y_test = train_test_split(self.X, self.y, test_size=0.3, random_state=42)

    def train_models(self):
        self.models['Gaussian Naive Bayes'] = GaussianNB()
        self.models['Multinomial Naive Bayes'] = MultinomialNB()
        self.models['Bernoulli Naive Bayes'] = BernoulliNB()
        self.models['Complement Naive Bayes'] = ComplementNB()

        for name, model in self.models.items():
            model.fit(self.X_train, self.y_train)
            score = model.score(self.X_test, self.y_test)
            print("{} Accuracy: {:.2f}%".format(name, score * 100))
    
            plot_decision_regions(self.X_train, self.y_train, clf=model, legend=2)
            plt.xlabel('sepal length (cm)')
            plt.ylabel('sepal width (cm)')
            plt.title('{} Decision Boundary'.format(name))
            plt.show()


if __name__ == '__main__':
    nb_models = NaiveBayesModels()
    nb_models.load_data()
    nb_models.train_models()
```
