## MSE function

```python
def mean_squared_error(y_true, y_pred):
    mse = np.mean((y_true - y_pred) ** 2)
    return mse
```

```python
y_true = np.array([1, 2, 3, 4, 5])
y_pred = np.array([1.2, 1.8, 2.9, 4.2, 4.8])

error = mean_squared_error(y_true, y_pred)
print(f'Mean Squared Error: {error:.3}')
```

## Plot loss Function

We have this formula:
- (y_hat -  y) ** 2 = (b1*x + b0 - y) ** 2
- In this formula, x and y are treated as constants, while **b0** and **b1** are variables.
- for simple explanation, we omit b0. so we have b1 and we want to plot different values of b1. formula changes to:
- (b1*x + 0 - y) = (b1*x - y)

```python
x = np.array([10, 8, 30, 4, 8])
y = np.array([2, 4, 5, 6, 2])

beta1_range = np.linspace(-10, 10, 1000)

mse_values = []
for b1 in beta1_range:
    y_pred = b1 * x
    mse = np.mean((y_pred - y) ** 2)
    mse_values.append(mse)

plt.plot(beta1_range, mse_values)
plt.xlabel('Slope (beta1)')
plt.ylabel("Mean Squared Error")
plt.title('MSE function for Linear Regression');
```

```python
min_loss_idx = np.argmin(mse_values)
loss_min = mse_values[min_loss_idx]

plt.plot(beta1_range, mse_values)
plt.scatter(beta1_range[min_loss_idx], loss_min, color='r');
```

- We should find "beta1" which is minimize the above function and this is an optimization problem

# Optimizer

In machine learning, an optimizer is an algorithm that is used to adjust the parameters of a model in order to minimize the loss function. \
The loss function is a measure of how well the model is performing on the training data, and the goal of the optimizer is to find the values of the model's parameters that minimize this loss function.

Here's an example of using Gradient Descent (GD) to find the minimum of the function y=0.2x^2:

```python
x = np.linspace(-10, 10, 100)
y = (0.2*x) **2

plt.title('$y=0.2x^2$')
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y');
```

How GD works:
1. Define the function and its derivative:

   y = 0.2x^2 \
   dy/dx = 0.4x

2. Initialize the value of x to a random value, and set the learning rate alpha to a small value (e.g., 0.1) and the maximum number of iterations to a large value (e.g., 1000).

   x = 5            # initial value of x \
   alpha = 0.1      # learning rate \
   max_iter = 1000  # maximum number of iterations 

3. Apply GD iteratively to update the value of x, using the formula:

   x := x - alpha * dy/dx

   where := means the value of x is updated.

   This formula updates x in the direction of the **negative gradient** of the function, which corresponds to moving towards the minimum.

4. Repeat step 3 until convergence or the maximum number of iterations is reached. Convergence can be checked by monitoring the change in the value of x at each iteration. When the change becomes very small (below a threshold), the algorithm can be considered to have converged.

## Implement function and gradient function

```python
def func(x):
    return (0.2*x)**2

def dfunc(x):
    return 0.4*x
```

## Plot function

```python
x = np.linspace(-10, 10, 100)
y = (0.2*x) **2

plt.title('$y=0.2x^2$')
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y');
```

## Implement GD algorithm

```python
x = 8
alpha = 0.1
max_iter = 1000
```

```python
for i in range(max_iter):
    dx = dfunc(x)
    x_new = x - alpha * dx
    x = x_new

print('Minimum value of y', func(x))
print('Value of x at minumum', x)
```

```python
x_value = x
min_y = func(x)
```
