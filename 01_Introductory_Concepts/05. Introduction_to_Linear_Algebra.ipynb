## 1. Plotting Polynomial Vector, Addition, and Scalar Multiplication:

```python
import numpy as np  
import matplotlib.pyplot as plt  

# Define the polynomial functions  
def f(x):  
    return x**4 - 2*x + 3  

def g(x):  
    return -2*x**3 + x**2 - 3*x  


# Scalar   
scalar = 3  

# Define the polynomial functions after scalar multiplication  
def h(x):  
    return scalar * f(x)  

# Generate x values  
x = np.linspace(-2, 2, 400)  

# Compute y values for the original polynomials  
y_f = f(x)  
y_g = g(x)  
y_scalar = h(x)           # Scalar multiplied function  
y_sum = f(x) + g(x)       # Polynomial addition  
y_prod = f(x) * g(x)      # Polynomial production  

# Create the plot  
plt.figure(figsize=(8, 6))  

# Plot the original polynomial functions  
plt.plot(x, y_f, label=r'$f(x) = x^4 - 2x + 3$', color='blue')  
plt.plot(x, y_g, label=r'$g(x) = -2x^3 + x^2 - 3x$', color='green')  

# Plot the added, scalar, and multiplied polynomial  
plt.plot(x, y_sum, label=r'$f(x) + g(x)$', color='orange')  
plt.plot(x, y_scalar, label=r'$3 \cdot f(x)$', color='red')  
plt.plot(x, y_prod, label=r'$f(x)*g(x)$', color='k') 

# Set plot details  
plt.title('Polynomial Functions, Addition, and Scalar Multiplication')  
plt.xlabel('x')  
plt.ylabel('y')  
plt.axhline(0, color='black', lw=0.5, ls='--')  # x-axis  
plt.axvline(0, color='black', lw=0.5, ls='--')  # y-axis  
plt.grid()  
plt.legend()  
plt.ylim(-10, 50)  # Set y limits for better visualization  
plt.xlim(-2, 2)  
plt.show()
```

## 2. Implementations of Fundamental Vector Operations in Linear Algebra

```python
import numpy as np  

# 1. Addition  
def vector_add(v1, v2):  
    """Adds two vectors."""  
    return np.array(v1) + np.array(v2)  

# 2. Subtraction  
def vector_subtract(v1, v2):  
    """Subtracts vector v2 from vector v1."""  
    return np.array(v1) - np.array(v2)  

# 3. Scalar Multiplication  
def scalar_multiply(scalar, vector):  
    """Multiplies a vector by a scalar."""  
    return scalar * np.array(vector)  

# 4. Dot Product (Scalar Product)  
def dot_product(v1, v2):  
    """Calculates the dot product of two vectors."""  
    return np.dot(np.array(v1), np.array(v2))  

# 5. Cross Product (Vector Product)  
def cross_product(v1, v2):  
    """Calculates the cross product of two vectors."""  
    return np.cross(np.array(v1), np.array(v2))  

# 6. Magnitude (Length)  
def vector_magnitude(vector):  
    """Calculates the magnitude of a vector."""  
    return np.linalg.norm(np.array(vector))  

# 7. Normalization  
def normalize_vector(vector):  
    """Normalizes a vector."""  
    magnitude = vector_magnitude(vector)  
    return np.array(vector) / magnitude if magnitude else np.array(vector)  

# 8. Angle Between Vectors  
def angle_between(v1, v2):  
    """Calculates the angle in degrees between two vectors."""  
    v1_u = normalize_vector(v1)  
    v2_u = normalize_vector(v2)  
    dot = dot_product(v1_u, v2_u)  
    angle_rad = np.arccos(dot)  
    return np.degrees(angle_rad)  

# Example Usage  
v1 = [3, 4, 0]  
v2 = [4, -3, 0]  

print('v1:', v1)
print('v2:', v2)
print()

print("Addition:", vector_add(v1, v2))  
print("Subtraction:", vector_subtract(v1, v2))  
print("Scalar Multiplication (2 * v1):", scalar_multiply(2, v1))  
print("Dot Product:", dot_product(v1, v2))  
print("Cross Product:", cross_product(v1, v2))  
print("Magnitude of v1:", vector_magnitude(v1))  
print("Normalized v1:", normalize_vector(v1))  
print("Angle Between Vectors:", angle_between(v1, v2), "degrees")
```
